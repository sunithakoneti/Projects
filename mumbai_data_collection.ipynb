```python
## importing required libraries
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
## Requesting for the data
url = "https://www.cars24.com/buy-used-car?f=make%3A%3D%3Aford%3Bmodel%3Ain%3Aecosport&sort=bestmatch&serveWarrantyCount=true&gaId=1400904043.1678979027&storeCityId=2378&pinId=400001"
page = requests.get(url)
soup = BeautifulSoup(page.text)

## make
make = []
for i in soup.find_all("h2",class_="_2lmIw"):
    a = i.text
    make.append(a.split()[1])

## Model
model = []
for i in soup.find_all("h2",class_="_2lmIw"):
    a = i.text.split()
    model.append(a[2])  
model
## year of the model
year = []
for i in soup.find_all("h2",class_="_2lmIw"):
    a = i.text.split()
    year.append(a[0])

# soup.find_all("ul",class_="_1hOnS")

# Series
series = []
for i in soup.find_all("ul",class_="_1hOnS"):
    a = i.text
    b = re.findall(r"(.+?)(Automatic|Manual)",a)
    series.append(b[0][0])
    
## Transmission
transmission = []
for i in soup.find_all("ul",class_="_1hOnS"):
    a = i.text
    b = re.findall(r"(.+?)(Automatic|Manual)",a)
    transmission.append(b[0][1])

# soup.find_all("ul",class_="_13yb6")

## kilometers
kilometers = []
for i in soup.find_all("ul",class_="_13yb6"):
    a = i.text.split("km")
    kilometers.append(a[0].replace(",",""))

## Owner
owner = []
for i in soup.find_all("ul",class_="_13yb6"):
    a = i.text.split("km")
    owner.append(a[1].split("Owner")[0])


## Fuel

fuel=[]
for i in soup.find_all("ul",class_="_13yb6"):
    a=i.text
    fuel.append(a.split("Owner")[1][:-5]) 

# soup.find_all("div",class_="_1Em-A")

## EMI
emi = []
for i in soup.find_all("div",class_="_1Em-A"):
    a = i.text.split("/")
    emi.append(a[0].replace("₹","").replace(",",""))

# soup.find_all("div",class_="_18ToE")

## Price
price = []
for i in soup.find_all("div",class_="_18ToE"):
    a = i.text.split("₹")
    price.append(a[1].replace(",",""))

# #print(make,model,series,transmission,emi,fuel,price,owner,kilometers,year) 
# print(len(make)) 
# print(len(model))
# print(len(series))
# print(len(transmission))
# print(len(emi))
# print(len(fuel))
# print(len(price))
# print(len(owner))
# print(len(kilometers))
# print(len(year))

#city
city=["Mumbai"]*20
# state
state = ["Maharashtra"]*20
# Mileage
mileage = [9]*20
# Seating
seating = [5]*20

d={"state":state,"city":city,"year":year,"make":make,"model":model,"series":series,"transmission":transmission,"fuel":fuel,"owner":owner,"emi":emi,
   "kilometers":kilometers,"mileage":mileage,"seating":seating,"price":price}

used_cars_mumbai_ford_ecosport_data_collection = pd.DataFrame(d)
used_cars_mumbai_ford_ecosport_data_collection.head()
#used_cars_mumbai_ford_ecosport_data_collection.to_csv("mumbai_ford_ecosport.csv")
```
